```python
nifty_financial_services_25_50_1 = pd.read_csv("/Users/neerajsmacbook/Downloads/Data Visualization Project/Indices Data (OHLC)/Historical Index Data/NIFTY FINANCIAL SERVICES 25_50_Historical_PR_01012015to31122024.csv")
nifty_financial_services_25_50_2 = pd.read_csv("/Users/neerajsmacbook/Downloads/Data Visualization Project/Indices Data (OHLC)/P:E, P:B & Dividend Ratio/NIFTY FINANCIAL SERVICES 25_50_Historical_PE_PB_DIV_Data_01012015to31122024.csv")

nifty_financial_services_25_50 = pd.merge(nifty_financial_services_25_50_1, nifty_financial_services_25_50_2, on = 'Date', how = 'inner')

nifty_financial_services_25_50.drop('IndexName', axis = 1, inplace = True)

from datetime import datetime

nifty_financial_services_25_50["Date"] = pd.to_datetime(nifty_financial_services_25_50["Date"], format = "%d %b %Y").dt.strftime("%d/%m/%Y")

nifty_consumer_durables.replace('-', np.nan, inplace=True)

nifty_financial_services_25_50['Open'] = pd.to_numeric(nifty_financial_services_25_50['Open'], errors='coerce')
nifty_financial_services_25_50['High'] = pd.to_numeric(nifty_financial_services_25_50['High'], errors='coerce')
nifty_financial_services_25_50['Low'] = pd.to_numeric(nifty_financial_services_25_50['Low'], errors='coerce')
nifty_financial_services_25_50['P/E'] = pd.to_numeric(nifty_financial_services_25_50['P/E'], errors='coerce')
nifty_financial_services_25_50['P/B'] = pd.to_numeric(nifty_financial_services_25_50['P/B'], errors='coerce')

mean_value_open = nifty_financial_services_25_50['Open'].mean()
mean_value_high = nifty_financial_services_25_50['High'].mean()
mean_value_low = nifty_financial_services_25_50['Low'].mean()
mean_value_p_e = nifty_financial_services_25_50['P/E'].mean()
mean_value_p_b = nifty_financial_services_25_50['P/B'].mean()

nifty_financial_services_25_50['Open'].fillna(mean_value_open, inplace=True)
nifty_financial_services_25_50['High'].fillna(mean_value_high, inplace=True)
nifty_financial_services_25_50['Low'].fillna(mean_value_low, inplace=True)
nifty_financial_services_25_50['P/E'].fillna(mean_value_p_e, inplace=True)
nifty_financial_services_25_50['P/B'].fillna(mean_value_p_b, inplace=True)
```
