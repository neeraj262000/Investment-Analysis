```python
import yfinance as yf
import pandas as pd
import numpy as np
```

```python
INDICES = ["^NSEI", "^NSEBANK", "^CNXIT", "^CNXAUTO", "^CNXFMCG", "^CNXMETAL", "^CNXREALTY", "^CNXENERGY", "^CNXMEDIA", "^CNXPSUBANK", "^BSESN"]

for i in INDICES:
    index_data = yf.download(i, start='2015-01-01', end='2024-12-31')
    
    variable_name = i.replace('^', '')

    if isinstance(index_data.columns, pd.MultiIndex):
        index_data.columns = index_data.columns.droplevel(1)

    index_data['Change_Percentage_Indices'] = index_data['Close'].pct_change() * 100
    index_data['Open'] = index_data['Open'].round(2)
    index_data['High'] = index_data['High'].round(2)
    index_data['Low'] = index_data['Low'].round(2)
    index_data['Close'] = index_data['Close'].round(2)
    index_data['Change_Percentage_Indices'] = index_data['Change_Percentage_Indices'].round(2)
    
    globals()[variable_name] = index_data
    
    
NSEI["Index_name"] = "NIFTY 50"
NSEBANK["Index_name"] = "NIFTY BANK"
CNXIT["Index_name"] = "NIFTY IT"
CNXAUTO["Index_name"] = "NIFTY AUTO"
CNXFMCG["Index_name"] = "NIFTY FMCG"
CNXMETAL["Index_name"] = "NIFTY METAL"
CNXREALTY["Index_name"] = "NIFTY REALTY"
CNXENERGY["Index_name"] = "NIFTY ENERGY"
CNXMEDIA["Index_name"] = "NIFTY MEDIA"
CNXPSUBANK["Index_name"] = "NIFTY PSU BANK"
BSESN["Index_name"] = "SENSEX"

INDICES = pd.concat([NSEI, NSEBANK, CNXIT, CNXAUTO, CNXFMCG, CNXMETAL, CNXREALTY, CNXENERGY, CNXMEDIA, CNXPSUBANK, BSESN], ignore_index=False)

INDICES = INDICES[['Index_name'] + [col for col in INDICES.columns if col != 'Index_name']]

```

```python
#Daily Return Percentage
INDICES['Daily_Return_%'] = ((INDICES['Close'] - INDICES['Open']) / INDICES['Open']) * 100
INDICES['Daily_Return_%'] = INDICES['Daily_Return_%'].round(2)
INDICES['Daily_Return_%'] = INDICES['Daily_Return_%'].map(lambda x: f'{x:.2f}%')

INDICES['Change_Percentage_Indices'] = INDICES['Change_Percentage_Indices'].map(lambda x: f'{x:.2f}%')

#Volatility
INDICES['Volatility'] = ((INDICES['High'] - INDICES['Low']) / INDICES['Open']) * 100
INDICES['Volatility'] = INDICES['Volatility'].round(2)
```
