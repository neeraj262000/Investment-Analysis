```python
start_date = "2015-01-01"
end_date = "2024-12-31"
div_merged = []


# Create an empty dictionary to store dataframes
dividend_data = {}

# Loop through each stock and download dividends
for ticker in NIFTY_50:
    
    # Download dividend data
    dividends_data = yf.Ticker(ticker).history(start=start_date, end=end_date, actions=True)

    dividends_data["Stock"] = ticker.replace('.NS', "")
    
    variable_name = ticker.replace('.NS', '')

    
    div_merged.append(dividends_data)
    
    final_df = pd.concat(div_merged, ignore_index=False)
    
        
    globals()[variable_name] = dividends_data
```
